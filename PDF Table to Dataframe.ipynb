# Extract Harding Loevner tables
- Extract sector and country weights from Harding Loevner Global Equity quarterly reports

```python
import pandas as pd
import tabula
from IPython.display import display_html, HTML

from os import listdir
column_name = ['HL Glbl', 'MSCI ACWI']
in_directory_1 = '/Users/ebellord/Desktop/global_quarterly_reports/'
areas_1= [[14.74, 237.98, 191.59, 196.4], [17, 503, 184, 177]]

in_directory_2 = '/Users/ebellord/Desktop/global_quarterly_reports2/'
areas_2 = [[46, 250, 182, 175], [48, 491, 183, 154]]
# areas_2 = [[46, 250, 182, 172], [48, 491, 183, 154]]

in_directory_3 = '/Users/ebellord/Desktop/global_quarterly_reports3/'
areas_3 = [[45, 260, 185, 189], [47, 487, 183, 165]]

in_directory_4 = '/Users/ebellord/Desktop/global_quarterly_reports4/'
areas_4 = [[45, 250, 185, 182], [47, 487, 183, 155]]
# areas_3 = [[45, 260, 185, 189], [47, 487, 183, 165]]
def process_pdf(in_directory, areas):
    sector_area = areas[0]
    country_area = areas[1]
    all_sectors = []
    all_regions = [] 
    for item in listdir(in_directory):  # loop through items in dir
        if item.endswith('.pdf'):
            path = in_directory + item
            def read_table(left, top, width, height):
                right = left + width
                bottom = top + height
                df = tabula.read_pdf(path, pages = '1',
                                 multiple_tables = False,
                                 guess = False,
                                 stream = True,
                                 area = [top, left, bottom, right])
                print(df.columns)
                # Remove numbers from column names before setting index
                df[df.columns[0]] = df[df.columns[0]].str.replace('\d+', '')
                df[df.columns[0]] = df[df.columns[0]].str.lower()
                df = df.set_index(df.columns[0])
                df = df.replace('–', 0)
                df = df.replace('-', 0)
                df = df.replace('—', 0)
                return df

            left, top, width, height = sector_area
            sector = read_table(left, top, width, height)
            sector.name = item.strip('-Report.pdf')
            sector.name = sector.name.strip('GE-')
            sector.columns = column_name
            print(sector.name)
            all_sectors.append(sector)
            left, top, width, height = country_area
            region = read_table(left, top, width, height)
            region.name = item.strip('-Report.pdf')
            region.name = region.name.strip('GE-')
            region.columns = column_name
            all_regions.append(region)
    return all_regions, all_sectors

group1, group2, group3, group4 = [process_pdf(directory, areas) for directory, areas in zip([in_directory_1,
                                                                              in_directory_2,
                                                                              in_directory_3,
                                                                              in_directory_4],
                                                                              [areas_1,
                                                                              areas_2,
                                                                              areas_3, 
                                                                              areas_4])]
# group2 = [process_pdf(directory, areas) for directory, areas in zip([in_directory_2],
#                                                                               [areas_2])]
#     # display_html(sector.style.set_table_attributes("style='display:inline'")._repr_html_() + 
    #              region.style.set_table_attributes("style='display:inline'")._repr_html_(), raw=True)
```

```python
group3_sectors = []
for item in group3[1]:
    name = item.name
    idx = item.index.str.strip('..')
    idx = idx.str.strip('.-')
    idx = idx.str.strip('.–')
    idx.name = 'Sector'
    item = item.set_index(idx)
    item.name = name
    group3_sectors.append(item)

group3_regions = []
for item in group3[0]:
    name = item.name
    idx = item.index.str.strip('..')
    idx = idx.str.strip('.-')
    idx = idx.str.strip('.–')
    idx.name = 'Region'
    item = item.set_index(idx)
    item.name = name
    group3_regions.append(item)
```

```python
group1_regions = group1[0]
group1_sectors = group1[1]

regions = group1_regions + group2_regions + group3_regions + group4_regions
sectors = group1_sectors + group2_sectors + group3_sectors + group4_sectors
```

```python
temp_list = []
for item in regions:
    item['Active']=item['HL Glbl'].astype(float) - item['MSCI ACWI'].astype(float)
    temp_list.append(item)
regions = temp_list   
active_regions = [item['Active'] for item in regions]  
names = [item.name for item in regions]    
active_regions = [item.rename(name) for item, name in zip(active_regions, names)]   
active_region = pd.concat(active_regions, axis=1, sort=False).transpose()
active_region = active_region.set_index(pd.to_datetime(active_region.index)+ pd.offsets.QuarterEnd(0))
active_region = active_region.sort_index()

temp_list = []
for item in sectors:
    item['Active']=item['HL Glbl'].astype(float) - item['MSCI ACWI'].astype(float)
    temp_list.append(item)
sectors = temp_list   
active_sectors = [item['Active'] for item in sectors]  
names = [item.name for item in sectors]    
active_sectors = [item.rename(name) for item, name in zip(active_sectors, names)]   
active_sectors = pd.concat(active_sectors, axis=1, sort=False).transpose()
active_sectors = active_sectors.set_index(pd.to_datetime(active_sectors.index)+ pd.offsets.QuarterEnd(0))
active_sectors = active_sectors.sort_index()
```
